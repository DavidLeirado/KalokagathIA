```python
# https://github.com/cdpierse/transformers-interpret
from transformers import AutoTokenizer
from transformers_interpret import SequenceClassificationExplainer
import torch
import json
from tqdm import tqdm

tokenizer = AutoTokenizer.from_pretrained('nlpaueb/bert-base-greek-uncased-v1')
interpreter = SequenceClassificationExplainer(model, tokenizer)
```

```python
with open('../input/val-data/labels_encoder.json', 'r') as f:
    json_dict = f.read()
    label_encoder = json.loads(json_dict)
print(label_encoder)
```

```python
class TextPredictor():
    def __init__(self, df, text_name, author, model=model, tokenizer=tokenizer, encoder=label_encoder, flag='interpreter'):
        self.df = df
        self.text = text_name
        self.author = author
        self.model = model
        self.tokenizer = tokenizer
        self.encoder = encoder
        self.chunks = self.fragment()
        self.flag = flag

    def fragment(self):
        text_list = self.df[self.df.Obra == self.text].Texto.to_list()
        text = ' '.join(text_list)
        #print(text)
        splitted_text = text.split()
        n = 100
        text_chunks = [' '.join(splitted_text[i:i+n]) for i in range(0,len(splitted_text),n)]
        return text_chunks

    def predict(self):
        predictions = self.init_prediction_dict()
        for text in tqdm(self.chunks):
            
            if self.flag = 'interpreter':
                # interpreter = SequenceClassificationExplainer(self.model, self.tokenizer)
                prediction = interpreter(text)
                #print(f'Prediction: {prediction}\n\n')
                predicted_label=int(interpreter.predicted_class_name.split('_')[-1])
                #print(predicted_label)
                predicted_accuracy = sum([i for _,i in prediction])
                #print(predicted_accuracy)
                #interpreter.visualize()
                for autor,label in self.encoder.items():
                    if label == predicted_label:
                        predictions[autor] = predictions[autor].append(predicted_accuracy)
                    else:
                        predictions[autor] = predictions[autor].append(None)
                        
            elif self.flag == 'model':
                tokenized_text = tokenizer(text, return_tensors = 'pt')
                prediction = softmax(model(**tokenized_text).logits.detach().numpy()[0])
                
                for autor, label in self.encoder.items():
                    predictions[autor] = predictions[autor].append(prediction[label])
    
    
        self.show_predictions(predictions)
                

    def show_predictions(self, predictions):
        print(f'Text: {self.text}')
        print('True author: '+self.author)
        predicted_autor, confidence = self.confidence(predictions)
        print('\tPrediction: {}\n\tConfidence: {} %'.format(predicted_autor,str(confidence)))

    def init_prediction_dict(self):
        pred_dict = {}
        for autor in self.encoder.keys():
            pred_dict[autor] = []
        return pred_dict
    
    def confidence(self, predictions, top=3):
        '''
        Modificar para nuevo dict: {Autor: []}
        '''
        pred_dict = {}
        predictions_temp = predictions
        
        for _ in range(top):
            predicted_autor = max(predictions_temp, key=predictions.get)    
            confidence = predictions[predicted_autor]/sum(predictions.values()) * 100
            
            pred_dict[predicted_autor] = confidence
            del predictions_temp[predicted_autor]
        
        return pred_dict
```
